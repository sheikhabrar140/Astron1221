# Parallax 
Parallax denotes the tiny shift in a star's apparent position when viewed from two different points in space. Let us look into a basic calculation of parallax. We will be using the texbook formula $d(\mathrm{pc}) = \frac{1}{p(\mathrm{arcsec})}$

```python
import math
parallax_mas = 742  # Alpha Centauri's parallax in milliarcseconds
distance_pc = 1000 / parallax_mas # Parallax in mas, so multiply by 1000 for parsecs
print(f"Alpha Centauri is {distance_pc:.2f} parsecs away")
```

### **Parallax Error Propagation.**   
The basic formula is $d(\mathrm{pc}) = \frac{1}{p(\mathrm{arcsec})}$. Now suppose the parallax is measured with some uncertainity. $p±σ_p$,where the $σ_p$ is standard deviation / measurement error in parallax.  
For a function $y=f(x)$   
If $x$ has an uncertainty $\sigma_x$, then for small errors, the uncertainty in $y$ is approximated by:

$\sigma_y \approx \left| \frac{dy}{dx} \right| \sigma_x$  
Applying it to $d=1/p$  
Derivative: $ \frac{d}{dp}\left(\frac{1}{p}\right) = -\frac{1}{p^2} $

Take absolute value for uncertainty: $ \sigma_d = \left| -\frac{1}{p^2} \right| \sigma_p = \frac{\sigma_p}{p^2} $




```python
parallax_mas = 747.17
parallax_mas_err = 0.61

parallax_as = parallax_mas/1000
parallax_as_err = parallax_mas_err/1000

distance_pc = 1/parallax_as
distance_pc_err = parallax_as_err/parallax_as**2 #See the explanation of error above

print(f"The distance from the Earth to the Alpha Centauri system is {distance_pc:.3f} \u00B1 {distance_pc_err:.3f}")
```
